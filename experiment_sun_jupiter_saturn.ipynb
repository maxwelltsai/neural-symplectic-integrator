```python
import rebound
import sys
import numpy as np
```

```python
sr = rebound.Simulation()
sr.units = {'yr', 'au', 'MSun'}
```

```python
sr.G
```

```python
sr.add('Sun')
sr.add('Jupiter')
sr.add('Saturn')
```

```python
sr.particles[1]
```

```python
sr.save('ic_sun_jupiter_saturn.bin')
```

```python
import abie
from wh import WisdomHolman
```

```python
t_end = 2000
```

```python
sim = WisdomHolman(CONST_G=sr.G)

for i, p in enumerate(sr.particles):
    print(p.m, p.x, p.y, p.z, p.vx, p.vy, p.vz)
    sim.particles.add(mass=p.m, pos=(p.x, p.y, p.z), vel=(p.vx, p.vy, p.vz))
```

```python
sim.output_file = 'sun_jupiter_saturn_nb.hdf5'
sim.integrator_warmup()
sim.h = 1.0
sim.acceleration_method = 'numpy'
sim.buf.recorder.set_monitored_quantities(['a', 'ecc', 'inc', 'x', 'y', 'z'])
sim.buf.recorder.start()
sim.integrate(t_end, nih=False)
sim.buf.flush()
```

```python
sim.stop()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(sim.buf.recorder.data['ecc'][:, 1:], alpha=0.5)
plt.show()
```

```python
from nn_models import MLP
from nih import NIH
import torch
from config import CONFIG
from utils import get_model_path, get_backbone


EXPERIMENT_DIR = '.'
sys.path.append(EXPERIMENT_DIR)

def load_model(config, device):
    output_dim = 1
    nn_model = get_backbone(config['backbone'])
    print(nn_model)
    model = NIH(config['input_dim'], differentiable_model=nn_model, device=device)

    path = get_model_path()
    print(path)
    model.load_state_dict(torch.load(path, map_location=torch.device(device)))
    return model

config = CONFIG
if torch.cuda.is_available():
  device = 'cuda'
else:
  device = 'cpu'
nih_model = load_model(config, device)
```

```python
sim2 = WisdomHolman(hnn=nih_model, CONST_G=sr.G)
# sim.integrator = 'WisdomHolman'

for i, p in enumerate(sr.particles):
    print(p.m, p.x, p.y, p.z, p.vx, p.vy, p.vz)
    sim2.particles.add(mass=p.m, pos=(p.x, p.y, p.z), vel=(p.vx, p.vy, p.vz))
```

```python
sim2.output_file = 'sun_jupiter_saturn_nih.hdf5'
sim2.integrator_warmup()
sim2.h = 1.0
sim2.acceleration_method = 'numpy'
sim2.buf.recorder.set_monitored_quantities(['a', 'ecc', 'inc', 'x', 'y', 'z'])
sim2.buf.recorder.start()
sim2.integrate(t_end, nih=True)
sim2.buf.flush()
```

```python
%matplotlib inline 
plt.plot(sim2.buf.recorder.data['ecc'][:, 1:], alpha=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(2,5, figsize=(18,6))
data_nb = sim.buf.recorder.data
data_nih = sim2.buf.recorder.data
time = np.linspace(0, t_end, data_nih['x'].shape[0])

names = ['Sun', 'Jupiter', 'Saturn']

for i in range(1, data_nih['x'].shape[1]):
    # coordinates in the first column
    axes[0,0].plot(data_nih['x'][:,i], data_nih['y'][:,i], '.', label=names[i], alpha=0.6)
    axes[0,0].axis('equal')
    axes[0,0].set_ylabel('$y$')
    axes[0,0].legend()
    axes[1,0].plot(data_nb['x'][:,i], data_nb['y'][:,i], '.', label=names[i], alpha=0.6)
    axes[1,0].axis('equal')
    axes[1,0].set_xlabel('$x$')
    axes[1,0].set_ylabel('$y$')
    axes[1,0].legend()

    # semi-major in the second column
    axes[0,1].plot(time, data_nih['a'][:,i], label=names[i], alpha=0.6)
    axes[0,1].set_ylabel('$a$ [au]')
    axes[0,1].legend()
    axes[1,1].plot(time, data_nb['a'][:,i], label=names[i], alpha=0.6)
    axes[1,1].set_xlabel('$t$ [years]')
    axes[1,1].set_ylabel('$a$ [au]')
    axes[1,1].legend()
    
    # eccentricity in the second column
    axes[0,2].plot(time, data_nih['ecc'][:,i], label=names[i], alpha=0.6)
    axes[0,2].set_ylabel('$e$')
    axes[0,2].legend()
    axes[1,2].plot(time, data_nb['ecc'][:,i], label=names[i], alpha=0.6)
    axes[1,2].set_xlabel('$t$ [years]')
    axes[1,2].set_ylabel('$e$')
    axes[1,2].legend()
    
    # inclination in the second column
    axes[0,3].plot(time, np.degrees(data_nih['inc'][:,i]), label=names[i], alpha=0.6)
    axes[0,3].set_ylabel('$i$ [deg]')
    axes[0,3].legend()
    axes[1,3].plot(time, np.degrees(data_nb['inc'][:,i]), label=names[i], alpha=0.6)
    axes[1,3].set_xlabel('$t$ [years]')
    axes[1,3].set_ylabel('$i$ [deg]')
    axes[1,3].legend()
    
# energy drift in the second column
axes[0,4].plot(time, sim2.energy, alpha=0.6)
axes[0,4].set_ylabel('$dE/E_0$')
axes[1,4].plot(time, sim.energy, alpha=0.6)
axes[1,4].set_xlabel('$t$ [years]')
axes[1,4].set_ylabel('$dE/E_0$')

    
plt.tight_layout()
plt.savefig('sun_jupiter_saturn_%dyr.pdf' % t_end)
```
